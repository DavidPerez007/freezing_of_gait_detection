# Daphnet Freezing of Gait Dataset Loader

Este notebook implementa un sistema OOP para cargar y procesar el dataset Daphnet Freezing of Gait.

## Dataset Information
- **Fuente**: Laboratory for Gait and Neurodynamics, Tel Aviv Sourasky Medical Center
- **Sensores**: 3 acelerómetros 3D a 64 Hz
- **Ubicación**: Tobillo (shank), Muslo (thigh), Tronco (trunk)
- **Sujetos**: 10 pacientes con Parkinson's (S01-S10)
- **Anotaciones**: 0=No experimento, 1=Caminar sin freeze, 2=Freeze

## 6. Visualizaciones Básicas

```python
# Distribución de anotaciones
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Gráfico de barras
annotation_counts = df_daphnet['annotation_label'].value_counts()
axes[0].bar(annotation_counts.index, annotation_counts.values, color=['gray', 'green', 'red'])
axes[0].set_xlabel('Tipo de Anotación')
axes[0].set_ylabel('Número de Muestras')
axes[0].set_title('Distribución de Anotaciones')
axes[0].tick_params(axis='x', rotation=45)

# Gráfico de pastel
axes[1].pie(annotation_counts.values, labels=annotation_counts.index, 
            autopct='%1.1f%%', colors=['gray', 'green', 'red'])
axes[1].set_title('Proporción de Anotaciones')

plt.tight_layout()
plt.show()
```

```python
# Freezes por sujeto
plt.figure(figsize=(12, 6))
plt.bar(summary_by_subject.index, summary_by_subject['freeze_percentage'], 
        color='steelblue', edgecolor='black')
plt.xlabel('Subject ID')
plt.ylabel('Porcentaje de Freeze (%)')
plt.title('Porcentaje de Episodios de Freeze por Sujeto')
plt.xticks(summary_by_subject.index)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
